```python
#bai1
a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]

n1, n2 = len(a), len(b)
min_len = min(n1, n2)

# Lấy giá trị nhỏ nhất tại các vị trí tương ứng
res = [min(a[i], b[i]) for i in range(min_len)]

# Thêm phần dư của danh sách dài hơn
if n1 > n2:
    res.extend(a[min_len:])
else:
    res.extend(b[min_len:])

print("Kết quả Bai 1:", res)
```

```python
#bai2 a
n = 3
a = [[1, 2, 3],
     [4, 5, 6],
     [7, 8, 9]]

tong = 0
for i in range(n):
    for j in range(n - i): # Tối ưu vòng lặp j để chỉ chạy trong vùng tam giác
        tong += a[i][j]

print("a. Tổng tam giác trên đường chéo phụ:", tong)
```

```python
#bai2 b
m, n = 2, 3
a = [[-1, 2, -3],
     [4, -5, 6]]

for i in range(m):
    for j in range(n):
        a[i][j] = abs(a[i][j])

print("b. Ma trận sau khi lấy trị tuyệt đối:", a)
```

```python
#bai2 c
m, n = 2, 3
x = 99
a = [[1, 2, 3],
     [4, 5, 6]]

for i in range(m):
    for j in range(n):
        if a[i][j] % 2 == 0:
            a[i][j] = x

print(f"c. Ma trận sau khi thay số chẵn bằng {x}:", a)
```

```python
#bai2 d
m, n = 2, 2
a = [[2, 4],
     [6, 8]]

toan_chan = all(a[i][j] % 2 == 0 for i in range(m) for j in range(n))
print("d. Ma trận toàn chẵn:", "Đúng" if toan_chan else "Không")
```

```python
#bai2 e
n = 3
a = [[1, 2, 3],
     [2, 5, 4],
     [3, 4, 6]]

doi_xing = True
for i in range(n):
    for j in range(i + 1, n):
        if a[i][j] != a[j][i]:
            doi_xing = False
            break

print("e. Ma trận đối xứng:", "Đúng" if doi_xing else "Không")
```

```python
#bai2 f
n = 3
a = [[1, 0, 0],
     [0, 5, 0],
     [0, 0, 9]]

tang_dan = all(a[i][i] < a[i+1][i+1] for i in range(n - 1))
print("f. Đường chéo chính tăng dần:", "Đúng" if tang_dan else "Không")
```

```python
#bai2 g
n = 3
a = [[1, 2, 3],
     [4, 5, 6],
     [7, 8, 9]]

print("g. Các phần tử thuộc tam giác dưới đường chéo phụ:")
for i in range(n):
    for j in range(n):
        if i + j >= n - 1:
            print(a[i][j], end=" ")
    print()
```

```python
#bai2 h
n = 3
a = [[10, 0, 10],
     [0, 5, 0],
     [1, 0, 0]]

giam_dan = True
for i in range(n - 1):
    # Đường chéo phụ: hàng i cột (n-1-i)
    if a[i][n - 1 - i] <= a[i+1][n - 2 - i]:
        giam_dan = False
        break

print("h. Đường chéo phụ giảm dần:", "Đúng" if giam_dan else "Không")
```
