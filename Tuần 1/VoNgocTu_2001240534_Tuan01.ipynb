```python
#14d
import numpy as np
m= int(input("nhap m: "))
n= int(input("nhap n: "))
data = np.random.randn(m, n)
print(data)
Sum= np.sum(data, axis=1)
Max= np.argmax(Sum)
print(f"dong co tong lon nhat la: ",data[Max])
```

```python
#14e
import numpy as np
m= int(input("nhap m: "))
n= int(input("nhap n: "))
data = np.random.randn(m, n)
print(data)
Tich= np.prod(data, axis=1)
Min= np.argmin(Tich)
print(f"dong co tich nho nhat la: ",data[Min])
```
